```python
from bs4 import BeautifulSoup

wiki_url_response = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')

soup = BeautifulSoup(wiki_url_response.text, 'html.parser')

all_tables = soup.find('table',{'class':'wikitable sortable'}).tbody

rows = all_tables.find_all('tr')

columns = [v.text.replace('\n', '') for v in rows[0].find_all('th')]

df = pd.DataFrame()

for i in range(1, len(rows)):
    tds = rows[i].find_all('td')
    
    if len(tds) == 4:
        values = [tds[0].text, tds[1].text, tds[2].text.replace, tds[3].text]
    else:
        values = [td.text.replace('\n','') for td in tds]
        

    df = df.append(pd.Series(values, index=columns), ignore_index=True)
    
    df.to_csv(r'D:\Coursera\IBM Data Science\Labs\Python For DataScience\Python Final Assignment_SS\Converted CSV file Capstone\capstone.csv' )
   
```
